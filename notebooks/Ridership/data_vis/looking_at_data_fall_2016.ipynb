```python
import pandas as pd

df = pd.read_csv("mbta_ridership/MBTA_Bus_Ridership_by_Trip_Season_Route_Line_and_Stop/MBTA_Bus_Ridership_by_Trip_Season_Route_Line_and_Stop_Fall_2016.csv")

print(df.shape)       # rows, columns
print(df.columns)     # list of column names
print(df.head())      # first 5 rows
```

```python
df.describe(include="all")
```

```python
df['route_id'].unique()
```

```python
df['route_id'] = df['route_id'].astype(str)
print(df['route_id'].nunique(), "unique routes after cleanup")
```

```python
ridership_by_route = (
    df.groupby('route_id')['boardings']
    .sum()
    .reset_index()
    .sort_values(by='boardings', ascending=False)
)
print(ridership_by_route.head(10))
```

```python
sample_by_route = (
    df.groupby("route_id")["sample_size"]
    .sum()
    .reset_index()
    .sort_values(by="sample_size", ascending=False)
)

print(sample_by_route.head(10))   # routes with largest sample_size
print(sample_by_route.tail(10))   # routes with smallest sample_size
```

```python
print(df['route_id'].nunique(), "routes")
print(df['route_id'].unique()[:20])
```

```python
df['hour'] = pd.to_datetime(df['trip_start_time'], format="%H:%M:%S", errors='coerce').dt.hour
print(df['hour'].value_counts().sort_index())
print(f"\nNumber of invalid time values coerced to NaT: {df['hour'].isna().sum()}")
```

```python
def normalize_hour(t):
    h, m, s = map(int, t.split(":"))
    if h >= 24:
        h = h - 24
    return h

df['hour'] = df['trip_start_time'].astype(str).apply(normalize_hour)
```

```python
print(df['hour'].value_counts().sort_index())
```

```python
df.groupby('hour')['boardings'].sum().plot(kind='bar')
```
